### PraProses Data Hotspot MODIS 2001 - 2023

- Load Data 

```python
df_hotspot = pd.read_csv("D:/Dataset/dataset-firms-nasa/MODIS Indonesia 2023.csv", low_memory=False, parse_dates=["acq_date"])
df_hotspot
```

```python
df_hotspot_harian = df_hotspot.groupby(['acq_date']).size().reset_index(name='hotspot')
```

```python
print(df_hotspot_harian)
```

```python
# membuat frame
fig, ax = plt.subplots(figsize = (10,5))

# membuat time series plot
ax.plot(df_hotspot_harian["acq_date"], df_hotspot_harian["hotspot"], color="tab:blue", label="Hotspot di Indonesia 2001 - 2023", linewidth=2.5)

# membuat label-label
ax.set_title("", fontsize=14)
ax.set_xlabel("", fontsize=12)
ax.set_ylabel("", fontsize=12)
ax.legend(loc="best")
ax.grid(True)

# menampilkan plot
plt.show()
```

```python
df_hotspot_bulanan = df_hotspot.resample('M', on='acq_date').size().reset_index(name='hotspot')
```

```python
# membuat frame
fig, ax = plt.subplots(figsize = (10,5))

# membuat time series plot
ax.plot(df_hotspot_bulanan["acq_date"], df_hotspot_bulanan["hotspot"], color="tab:blue", label="Hotspot di Indonesia 2001 - 2023", linewidth=2.5)

# membuat label-label
ax.set_title("", fontsize=14)
ax.set_xlabel("", fontsize=12)
ax.set_ylabel("", fontsize=12)
ax.legend(loc="best")
ax.grid(True)

# menampilkan plot
plt.show()
```
