## Data MODIS Hotspot Sumatera Selatan 2001 - 2023

```python
# pustaka manipulasi data array
import numpy as np

# pustaka manipulasi data frame
import pandas as pd

# pustaka untuk visualisasi data
import matplotlib.pyplot as plt
```

- Load Dataset

```python
df_hotspot = pd.read_csv("D:/Dataset/dataset-firms-nasa/MODIS Sumsel 2023.csv", parse_dates=["acq_date"])
df_hotspot.info()
```

- Data Hotspot Sumatera Selatan Harian

```python
df_hotspot_harian = df_hotspot.groupby(['acq_date']).size().reset_index(name='hotspot')
df_hotspot_harian
```

```python
# membuat frame
fig, ax = plt.subplots(figsize = (10,5))

# membuat time series plot
ax.plot(df_hotspot_harian["acq_date"], df_hotspot_harian["hotspot"], color="tab:blue", label="Hotspot di Sumatera Selatan 2001 - 2023", linewidth=2.5)

# membuat label-label
ax.set_title("", fontsize=14)
ax.set_xlabel("", fontsize=12)
ax.set_ylabel("", fontsize=12)
ax.legend(loc="upper left")
ax.grid(True)

# menampilkan plot
plt.show()
```

- Data Hotspot Sumatera Selatan Bulanan

```python
df_hotspot_bulanan = df_hotspot.resample('M', on='acq_date').size().reset_index(name='hotspot')
df_hotspot_bulanan
```

```python
# membuat frame
fig, ax = plt.subplots(figsize = (10,5))

# membuat time series plot
ax.plot(df_hotspot_bulanan["acq_date"], df_hotspot_bulanan["hotspot"], color="tab:blue", label="Hotspot di Sumatera Selatan 2001 - 2023", linewidth=2.5)

# membuat label-label
ax.set_title("", fontsize=14)
ax.set_xlabel("", fontsize=12)
ax.set_ylabel("", fontsize=12)
ax.legend(loc="upper left")
ax.grid(True)

# menampilkan plot
plt.show()
```
